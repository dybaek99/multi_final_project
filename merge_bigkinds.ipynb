```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")

import pandas as pd

# 통합할 파일 목록
file_list = [
    './Text_Data/bigkinds/IT_과학/NewsResult_20170901-20200901.xlsx',
    './Text_Data/bigkinds/IT_과학/NewsResult_20200902-20240830.xlsx'
]

# 빈 리스트에 각 파일에서 읽은 데이터프레임을 저장
dataframes = []

# 각 파일을 읽어 데이터프레임으로 변환 후 리스트에 추가
for file in file_list:
    df = pd.read_excel(file, converters={'뉴스 식별자': str})  # '뉴스 식별자'는 해당 열의 실제 이름으로 대체
    dataframes.append(df)

# 데이터프레임을 하나로 통합
df_combined = pd.concat(dataframes)

# 중복된 행 제거
df_combined.drop_duplicates(inplace=True)

df_combined = df_combined.sort_values(by=['일자', 'URL'], ascending=[True, True])

# 그룹핑할 컬럼 리스트
group_columns = ['일자', '언론사', '기고자', '제목', '통합 분류1', '통합 분류2', '통합 분류3', 
                 '사건/사고 분류1', '사건/사고 분류2', '사건/사고 분류3', '인물', '위치', '기관', 
                 '키워드', '특성추출(가중치순 상위 50개)', '본문', 'URL', '분석제외 여부']

# 그룹핑할 컬럼들의 NULL 값을 빈 문자열로 대체
df_combined[group_columns] = df_combined[group_columns].fillna('')

# 각 그룹에서 뉴스 식별자 컬럼의 최소값을 유지
df_combined = df_combined.groupby(group_columns, as_index=False).agg({'뉴스 식별자': 'min'})

# '뉴스 식별자'를 맨 앞으로 이동
columns_order = ['뉴스 식별자'] + group_columns  # '뉴스 식별자'를 첫 번째로 이동시키고 나머지 컬럼을 뒤에 추가
df_combined = df_combined[columns_order]

# 통합된 데이터를 새로운 엑셀 파일로 저장
output_file = './Text_Data/bigkinds/삼성전자_IT_과학_20170901-20240830.xlsx'
df_combined.to_excel(output_file, index=False)

print(f"통합 및 중복 제거된 파일이 '{output_file}'로 저장되었습니다.")
```
